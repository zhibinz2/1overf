```python
import matplotlib.pyplot as plt

from hdf5storage import loadmat, savemat 

import numpy as np
```

```python
# load syn_pcorr
outdict=loadmat('pcorr_single.mat')
pcorr=outdict["pcorr"]
likelihood=outdict["likelihood"]
```

```python
# convert syn_pcorr to boolean
nonzerosmat=np.zeros((30,32,32))
for freq in range(30):			
	nonzerosmat[freq,:,:]=np.array(pcorr[freq,:,:],dtype=bool)
```

```python
# 1 plots to look number of edges at all freq 
sum_allfreq=np.zeros((30))
for freq in range(30):	
	sum_allfreq[freq]=(np.sum(nonzerosmat[freq,:,:])-32)/2
 

fig, ax = plt.subplots(1,1,figsize=(4, 4))
plt.plot (np.arange(30)+1,sum_allfreq)
plt.xlabel("frequency (Hz)")
plt.ylabel("# edges")
# plt.ylim(220,350)
plt.grid(); 
        
```

```python
# a selected a freq to examine the distribution
freq=10
fig, ax = plt.subplots(1,1,figsize=(6, 6))
plt.imshow(nonzerosmat[freq,:,:],vmin = 0,vmax = 1,cmap = 'Reds')
plt.title('freq ( ' + str(freq) + ' Hz )')
plt.colorbar()
```
